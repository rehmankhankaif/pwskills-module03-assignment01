<br>
Ans03: In Python, objects are classified into two main types: mutable and immutable. The primary difference between them is whether or not their state (or content) can be changed after they are created.<br>
<br>
1.0 Mutable Objects:<br>
Mutable objects can be changed after creation. This means you can alter their contents, add or remove elements, and make other modifications.<br>
<br>
Examples of Mutable Objects:<br>
<br>
1.1 Lists:<br>
<br>

<br>
1.2 Dictionaries:<br>
<br>

my_dict['c'] = 3 <br>

<br>
1.3 Sets:<br>
<br>

my_set.add(4) <br>
print(my_set)  <br>
<br>
<br>

Immutable objects cannot be changed after creation. If you need to modify an immutable object, you must create a new object with the desired changes.<br>
<br>
Examples of Immutable Objects:<br>
<br>
2.1 Strings:<br>
<br>

Ques06: How do conditional statements work in Python? Illustrate with examples<br>
<br>
Ans06: Conditional statements in Python allow us to execute different blocks of code based on certain conditions. The most commonly used conditional statements are if, elif (short for else if), and else. Here's a explanation with examples:<br>
<br>
1. if statement: This executes a block of code if a specified condition is true.<br>
x = 10<br>
if x > 5:<br>
    print("x is greater than 5")<br>
<br>
2. elif statement: This allows you to check multiple conditions. If the if condition is false, the elif condition is checked. You can have multiple elif statements.<br>
x = 5<br>
if x > 5:<br>
    print("x is greater than 5")<br>
elif x == 5:<br>
    print("x is equal to 5")<br>
<br>
3. else statement: This executes a block of code if none of the preceding if or elif conditions are true.<br>
x = 2<br>
if x > 5:<br>
    print("x is greater than 5")<br>
elif x == 5:<br>
    print("x is equal to 5")<br>
else:<br>
    print("x is less than 5")<br>
<br>
4. Nested conditional statements: You can also use conditional statements inside other conditional statements.<br>
x = 10<br>
y = 20<br>
if x > 5:<br>
    if y > 15:<br>
        print("x is greater than 5 and y is greater than 15")<br>
    else:<br>
        print("x is greater than 5 but y is not greater than 15")<br>
else:<br>
    print("x is not greater than 5")<br>

Ques07: Describe the different types of loops in Python and their use cases with examples<br>
<br>
Ans07: In Python, there are primarily two types of loops: for loops and while loops. Each of these loops allows you to execute a block of code multiple times, but they are used in different scenarios.<br>
<br>
1. for loop: This loop is used for iterating over a sequence (such as a list, tuple, string, or range). It is commonly used when you know the number of iterations in advance.<br>
fruits = ["apple", "banana", "cherry"]<br>
for fruit in fruits:<br>
    print(fruit)<br>
<br>
for i in range(5):<br>
    print(i)<br>
<br>
<br>
2. while loop: This loop is used when you want to repeat a block of code as long as a condition is true. It is commonly used when the number of iterations is not known in advance.<br>
count = 0<br>
while count < 5:<br>
    print(count)<br>
    count += 1<br>
<br>
<br>
3. Nested loops: You can also use loops inside other loops. This is useful for working with multi-dimensional data structures like lists of lists.<br>
matrix = [<br>
    [1, 2, 3],<br>
    [4, 5, 6],<br>
    [7, 8, 9]<br>
]<br>
for row in matrix:<br>
    for element in row:<br>
        print(element, end=" ")<br>
<br>
<br>
4. Break and continue statements: These statements control the flow of loops. The break statement terminates the loop, while the continue statement skips the current iteration and moves to the next one.<br>
for i in range(10):<br>
    if i == 5:<br>
        break<br>
    print(i)<br>
<br>
for i in range(10):<br>
    if i % 2 == 0:<br>
        continue<br>
    print(i)<br>
<br>
These loops are essential for executing repetitive tasks in your code. They allow you to process data, automate tasks, and implement algorithms efficiently.<br>
